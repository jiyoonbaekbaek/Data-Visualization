## 1. Loading the Raw Data -jy


```python
# Loading the data
raw_data=pd.read_csv('/content/kpop_idols_girl_groups.csv')

# runs all the data
print(raw_data.shape)

#runs the first 5 rows
raw_data.head()
```

## 4. Histograms - jy 

```python
# Example 1 

x = raw_data['Members'].values
sns.distplot(x, color = 'blue');
```

```python
# Example 2 - Investigating the distribution of Revenue, adding the mean

x = raw_data['Members'].values
sns.distplot(x, color = 'blue');

# Calculating the mean
mean = raw_data['Members'].mean()
print(mean)

#ploting the mean
#verticle line   - axvline(x ,y (0=),y (1=))
plt.axvline(mean, 0,1, color = 'red')
```

```python
# Example 3 

x=raw_data['Orig. Memb.'].values
sns.distplot(x, color = 'red');

# Calculating the mean
mean = raw_data['Orig. Memb.'].mean()
print(mean)

#ploting the mean

plt.axvline(mean, 0,1, color = 'blue')
```
