# Python Seaborn Tutorial for Beginners

## What is Seaborn?

- Python Data Visualization Library - based on MatPlotLib (see previous tutorials)
- Used for plotting statistical graphs, identifying trends, relationships & outliers
- In my opinion, Seaborn is easier & faster to use (less code) Vs MatPlotLib

# **JY** Example 2~4

```python
# Example 2 - Adding Categories
ax2=sns.lineplot(x='Orig. Memb.',y='Members',data=raw_data2[:50],hue='Company')
# By Company
```

```python
# Example 3 - By Promo with style
```

```python
# Example 4 - By Promo with style & Increase the size & Remove error bars

# increase the size
sns.set(rc={'figure.figsize':(12,10)})
ax2=sns.lineplot(x='Orig. Memb.',y='Members',data=raw_data2[:50],hue='Company',ci=False,style='Active')
```

```python
# Example 5 - By Promo with style & Increase the size & Remove error bars & adding markers 
ax2=sns.lineplot(x='Orig. Memb.',y='Members',data=raw_data2[:50],hue='Company',ci=False,style='Active',markers=True)
```
