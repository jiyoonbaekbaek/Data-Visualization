# 10 그리드 설정하기 : plt.grid(True)

```python
import matplotlib.pyplot as plt
import numpy as np

a=np.arange(0,2,0.2)
plt.plot(a,a**2)
plt.grid(True)
plt.show()
```

# 11 눈금 설정하기 : plt.xticks([...]),plt.yticks([...])

```python
a=np.arange(0,2,0.2)
plt.plot(a,a**2)
plt.xticks([0,1,2])
plt.yticks(np.arange(1,6))
plt.show()
```

## 눈금 스타일 설정하기 : tick_params( ) 
### 매개변수 설명 :
axis : 설정이 적용될 축 지정 
direction : in/ out 눈금이 안에 표시될지, 밖에 표시될지
length : 눈금의 길이, pad: 눈금과 레이블과의 거리
labelsize : 레이블의 크기, labelcolor : 레이블의 색상
top/bottom/left/right : 눈금이 표시될 위치 
width : 눈금의 너비, color : 눈금의 색상

```python
plt.rcParams['font.family'] = 'AppleGothic'
a=np.arange(0,2,0.2)
plt.plot(a,a**2)
plt.xticks([0,1,2],labels=['월','화','수'])
plt.yticks(np.arange(1,6))
plt.show()
```

# 12 타이틀 지정하기 : plt.title(loc,pad,fontdict)

```python
plt.rcParams['font.family'] = 'AppleGothic'
a=np.arange(0,2,0.2)
plt.plot(a,a**2)
plt.ylabel('유튜브 본 시간')
plt.xticks([0,1,2],labels=['월','화','수'])
plt.yticks(np.arange(1,6),labels=['1h','2h','3h','4h','5h'])
title_font={'fontsize':16,'fontweight':'bold'}
plt.title('I love youtube',loc='center',pad=20,fontdict=title_font)
plt.show()
```

# 13 수평선/수직선 표시하기 : 
axhline,hlines(y,xmin,xmax)
axvlines,vlines(x,ymin,ymax)

```python
plt.rcParams['font.family'] = 'AppleGothic'
a=np.arange(0,2,0.2)
print(a)
print(a**2)
plt.plot(a,a**2,marker='^')
plt.ylabel('유튜브 본 시간')
plt.xticks([0,1,2],labels=['월','화','수'])
plt.yticks(np.arange(1,6),labels=['1h','2h','3h','4h','5h'])
title_font={'fontsize':16,'fontweight':'bold'}
plt.title('I love youtube',loc='center',pad=20,fontdict=title_font)
plt.axhline(0.16,0,0.2,linestyle='--',linewidth=1)
plt.axvline(0.33,0,0.04,linestyle=':',linewidth=2)
plt.show()
```
