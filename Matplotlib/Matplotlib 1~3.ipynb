# 1. Matplotlib 기본 사용 ~ 3. Matplotlib 축 레이블 설정하기

```python
import matplotlib.pyplot as plt
import numpy as np
```

## plot 메소드 : 리스트의 값들이 y 값들이라고 가정하고, x 값들은 자동으로 만들어낸다

```python
plt.plot([1,2,3,4])
plt.show()
```

## 혹은 x값도 리스트로 넣어줄 수가 있음. plot([x값],[y값])

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

## axis 메소드 : 축의 범위 지정 [xmin,xmax,ymin,ymax]
### 'ro' 빨간색 원형 마커 , 'b-' 파란색 실선

```python
plt.plot([1,2,3,4],[1,4,9,16],'go')
plt.axis([0,6,0,20])
plt.show()
```

## 여러 개의 리스트를 plot 메소드에 한꺼번에 넣는 것도 가능

```python
x=np.array([1,2,3,4])
```

```python
plt.plot(x,x,'r-',x,x**2,'b-',x,x**3,'g-')
plt.show()
```

## 특정 데이터에서 플로팅할 수도 있음

```python
data_dict={'data_x':[1,2,3,4],'data_y':[2,5,9,18]}
```

```python
plt.plot('data_x','data_y',data=data_dict)
plt.show()
```

## xlabel 메소드 : x 라벨값 붙이기 , ylabel 메소드 : y 라벨값 붙이기
   

```python
plt.plot([1,2,3,4],[70,80,90,100])
plt.xlabel('Study')
plt.ylabel('Grades')
plt.show()
```

### labelpad 를 통해 패딩된 라벨값 지정

```python
plt.plot([1,2,3,4],[70,80,90,100])
plt.xlabel('Study',labelpad=15)
plt.ylabel('Grades',labelpad=20)
plt.show()
```

## 폰트 설정하기 font ={'family': ..., 'color' : ..., 'weight' : ... , 'size' : ... }

```python
font={'family':'fantasy',
    'color':'deeppink',
    'weight':'normal',
    'size':14}
```

```python
plt.plot([1,2,3,4],[70,80,90,100])
plt.xlabel('Study',fontdict=font)
plt.ylabel('Grades')
plt.show()
```

## 한글폰트 깨짐 해결하기 

```python
plt.rcParams['font.family'] = 'AppleGothic'
```

```python
plt.plot([1,2,3,4],[70,80,90,100])
plt.xlabel('공부 시간')
plt.ylabel('성적')
plt.show()
```
