# 4. 범례 표시하기 (범례란 ? 데이터의 종류를 표시하기 위한 텍스트)

## how ? plot 메소드의 라벨을 지정하고 legend 호출

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams['font.family'] = 'AppleGothic'
```

```python
plt.plot([1,2,3,4],[2,4,6,12],label='성적')
plt.xlabel('공부 시간')
plt.ylabel('성적')
plt.legend()
plt.show()
```

## 범례 위치 지정하기 : legend 메소드의 loc 옵션

```python
plt.plot([1,2,3,4],[2,4,6,12],label='성적')
plt.xlabel('공부 시간')
plt.ylabel('성적')
plt.legend(loc='lower right')
plt.show()
```

```python
# best, upper right, upper left, lower left, lower right, right, center left, center right,
# lower center, upper center, center
```

## 범례에 표시될 열 갯수 조정하기 : legend 메소드의 ncol 옵션

```python
plt.plot([1,2,3,4],[2,4,6,12],label='성적')
plt.plot([1,2,3,4],[3,5,12,15],label='성취감')
plt.xlabel('공부 시간')
plt.ylabel('얻는 것')
plt.legend(loc='best',ncol=2)
plt.show()
```

## 범례에 표시될 폰트 크기 조정하기 : legend 메소드의 fontsize 옵션

```python
plt.plot([1,2,3,4],[2,4,6,12],label='성적')
plt.plot([1,2,3,4],[3,5,12,15],label='성취감')
plt.xlabel('공부 시간')
plt.ylabel('얻는 것')
plt.legend(loc='best',ncol=2,fontsize=14)
plt.show()
```

## 범례 테두리 꾸미기 : legend 메소드의 frameon,shadow 등등의 옵션

```python
plt.plot([1,2,3,4],[2,4,6,12],label='성적')
plt.plot([1,2,3,4],[3,5,12,15],label='성취감')
plt.xlabel('공부 시간')
plt.ylabel('얻는 것')
plt.legend(loc='best',ncol=2,fontsize=14,frameon=True,shadow=True)
plt.show()
```

# 8. Matplotlib 그래프 영역 채우기 : 그래프의 특정 영역을 색상으로 채워서 강조

## how ? 
## fill_between() : 두 수평 방향의 곡선 사이를 채우기
## fill_betweenx() : 두 수직 방향의 곡선 사이를 채우기
## fill() : 다각형 영역을 채우기

```python
x=[1,2,3,4]
y=[70,75,85,100]
plt.plot(x,y,color='springgreen',label='성적')
plt.plot([1,2,3,4],[50,60,70,80],color='violet',label='성취감')
plt.xlabel('공부 시간')
plt.ylabel('얻는 것')
plt.axis('on')
plt.fill_betweenx(y[2:4],x[2:4],color='pink',alpha=0.5)
plt.legend(loc='best',ncol=2,fontsize=14,frameon=True,shadow=True)
plt.show()
```

```python
x=[1,2,3,4]
y=[70,75,85,100]
y2=[50,60,70,80]
plt.plot(x,y,color='springgreen',label='성적')
plt.plot(x,y2,color='violet',label='성취감')
plt.xlabel('공부 시간')
plt.ylabel('얻는 것')
plt.fill_between(x[1:3],y[1:3],y2[1:3],color='pink',alpha=0.5)
plt.legend(loc='best',ncol=2,fontsize=14,frameon=True,shadow=True)
plt.show()
```
