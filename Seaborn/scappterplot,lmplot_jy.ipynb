## Importing / Installing packages



```python
# Packages / libraries
import os #provides functions for interacting with the operating system
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# To install seaborn type "pip install seaborn" to the anaconda terminal

# import sys
# !conda list Check the packages installed
```

## 1. Loading the Raw Data


```python
# Loading the data

raw_data = pd.read_csv("flights.csv")
# runs all the data
raw_data

print(raw_data.shape)

#runs the first 5 rows
raw_data.head()
```

## 6. ScatterPlots

```python
raw_data.columns
```

```python
# Example 1 - Relationship between Marketing Spend and Revenue

ax = sns.scatterplot(x="distance", y="price", data=raw_data, color = 'green')
sns.set(rc={'figure.figsize':(12,10)})
```

```python
# Example 2 - Relationship between Marketing Spend and Revenue - changing color, hue & Style

ax = sns.scatterplot(x="distance", y="price", data=raw_data, color = 'green', hue = 'flightType', style = 'flightType')
```

```python
# Example 3 - Relationship between Marketing Spend and Revenue - changing color & hue - adding size

ax = sns.scatterplot(x="distance", y="price", data=raw_data, color = 'green', hue = 'flightType', size = 'price',
                    sizes=(20, 200))
```

```python
# Example 4 - Relationship between Marketing Spend and Revenue - changing color & hue - adding size - by day

ax = sns.scatterplot(x="time", y="price", data=raw_data, color = 'green', hue = 'flightType', size = 'price',
                    sizes=(20, 200))
```

## 7. lmPlots

```python
# it seems that lmplots is for linear regression
# Example 1 - Relationship between Marketing Spend and Revenue

ax = sns.lmplot(x="distance", y="price", data=raw_data, height=9)

# Notes:
# What is Linear Regression: It is a predictive statistical method for modelling the relationship between x (independent variable) & y (dependent V).
# How it works (cost function MSE): https://towardsdatascience.com/machine-learning-fundamentals-via-linear-regression-41a5d11f5220
```

```python
# Example 2 - Relationship between Marketing Spend and Revenue - changing color, hue & Style

ax = sns.lmplot(x="distance", y="price", data=raw_data, hue = 'flightType', ci= False, height=9, markers=["o", "x", "+"])
```

```python
# Example 3 - Relationship between Marketing Spend and Revenue - by column

ax = sns.lmplot(x="distance", y="price", data=raw_data, col = 'flightType', ci= False, height=5, 
                line_kws={'color': 'red'}, 
                scatter_kws={'color':'#ADC11E'})
```

```python
# Example 4 - Relationship between Marketing Spend and Revenue - by column - by day - add Jitter too
#           , x_jitter  y_jitter  .              .
ax = sns.lmplot(x="distance", y="price", data=raw_data, col = 'agency', ci= False, height=4, 
                line_kws={'color': '#031722'}, 
                scatter_kws={'color':'#1E84C1'},
               col_wrap=3,
               x_jitter=.3)
```
